```python
# Drop the columns of 'End_lat' & 'End_lng' from this dataset.
accidents_df.drop(['End_Lat', 'End_Lng'], axis=1, inplace=True)
```

```python
# Handling the missing data of 'Wind_Speed(mph)' & 'Precipitation(in)'
temp_df = accidents_df[['Wind_Speed(mph)', 'Precipitation(in)']]
temp_df.head()
```

```python
from sklearn.impute import KNNImputer

```

```python

knn_imputer = KNNImputer(n_neighbors=4)
```

```python
# Applying  KNN only on Wind_Speed column
temp_df[['Wind_Speed(mph)']] = knn_imputer.fit_transform(temp_df[['Wind_Speed(mph)']])

# Fill Weather_Conditions with mode
temp_df['Weather_Conditions'].fillna(temp_df['Weather_Conditions'].mode()[0], inplace=True)
```
